### Gradient Descent

````python
# prompt: TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

# ### Gradient Descent
def gradient_descent(derivative_func, initial_guess, learning_rate, precision, max_iters):
    # [REQUIRED]
    # Create variables for current x and previous x
    curr_x = initial_guess
    prev_x = 0
    # [REQUIRED]
    # Create a list to store the x values in each iteration
    x_list = [curr_x]

    # Iterate through the number of iterations
    for i in range(max_iters):
        # [REQUIRED]
        # Calculate the new x value based on the learning rate and the derivative of the function
        prev_x = curr_x
        curr_x -= learning_rate * derivative_func(prev_x)

        # [REQUIRED]
        # Append the current x value to the list of x values
        x_list.append(curr_x)

        # Check if the absolute difference between the current x and the previous x is less than the precision
        # If it is, then the algorithm has converged and we can stop
        if abs(curr_x - prev_x) < precision:
            break

    return curr_x, x_list
# Let's run the gradient descent algorithm for 100 iterations
# and see how the value of x changes.
local_min, list_x = gradient_descent(derivative_func=df, initial_guess=3, learning_rate=0.1, precision=0.0001, max_iters=100)
print("Local min occurs at:", local_min)
print("Number of steps:", len(list_x))
# Plot the gradient descent
plt.figure(figsize=[10, 5])
plt.xlim(-2, 3)
plt.ylim(0, 8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.plot(x_1, f(x_1), color='blue', linewidth=3, alpha=0.8)
plt.scatter(list_x, f(np.array(list_x)), color='red', s=100, alpha=0.6)

plt.show()
````

## Gradient Descent

````python
# Gradient Descent Function
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001, max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list

# Define the function and its derivative
def f(x):
    return x**2  # Example: f(x) = x^2

def df(x):
    return 2*x  # Derivative: f'(x) = 2x

# Perform gradient descent
min_x, x_list, slope_list = gradient_descent(df, initial_guess=3, multiplier=0.1, precision=0.0001)

# Results
print(f"Local minimum occurs at x = {min_x}")
print(f"f(x) at this point = {f(min_x)}")

# Visualization
import matplotlib.pyplot as plt
import numpy as np

# Data for the function
x = np.linspace(-3, 3, 100)
y = f(x)

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = x^2', color='blue')

# Plot the descent path
plt.scatter(x_list, [f(i) for i in x_list], color='red', label='Descent Path')
plt.title('Gradient Descent Optimization')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
````

#### Mean Squared Error: a cost function for regression problems

#### $$RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
#### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
#### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y - \hat{y} \big)^2 $$

````python
# Make sample data
x_5 = np.array([[0.1, 1.2, 2.4, 3.2, 4.1, 5.7, 6.5]]).transpose()
y_5 = np.array([1.7, 2.4, 3.5, 3.0, 6.1, 9.4, 8.2]).reshape(7, 1)

print('Shape of x_5 array:', x_5.shape)
print('Shape of y_5 array:', y_5.shape)
````

````python
# Challenge: Write a python function mse(y, y_hat) returns the MSE? Call the mse(y, y_hat)
# function and print out the MSE for the y_hat calculated above.

def mse(y, y_hat):
    #mse_calc = 1/7 * sum((y - y_hat)**2)
    #mse_calc = (1/y.size) * sum((y - y_hat)**2)
    mse_calc = np.average((y - y_hat)**2, axis=0)
    return mse_calc
````

#### Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

#### $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

#### $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

#### MSE & Gradient Descent

````python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
````

````python
multiplier = 0.01
thetas = np.array([2.9, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5)

for i in range(1000):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)

    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mse(y_5, thetas[0] + thetas[1]*x_5))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1]*x_5))
````
